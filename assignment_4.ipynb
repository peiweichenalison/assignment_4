```python
from bs4 import BeautifulSoup
import pandas as pd

html = open("diy.html")
soup = BeautifulSoup(html,'html.parser')

header = ['shopname','url']
data = []
for item in soup.find_all("div", attrs={"class":"txt"}): 
         shopname = item.find("div",attrs={"class":"tit"}).get_text().strip()
         url = item.find("div",attrs={"class":"tit"}).a.get('href')
         data.append((shopname,url))
            
df = pd.DataFrame(data,
   columns = header
)
df.to_csv('diy-shop.csv', sep='\t', encoding='utf-8-sig')#chinese website code
df
```

```python

 
import time
import logging
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
 
#设置用户名密码
username_  = "********"
password_  = "********"
 
#设置日志等级
logging.basicConfig(level=logging.INFO)
 
#打开浏览器
driver = webdriver.Chrome()
base_url = 'http://www.dianping.com/'
driver.get(base_url)
time.sleep(2)
 
# 右上登陆
login_btn =driver.find_element_by_xpath('//*[@id="top-nav"]/div/div[2]/span[1]/a[1]')
login_btn.click()
time.sleep(3)
 
# 选择账号登录
iframe =driver.find_element_by_xpath('//*[@id="J_login_container"]/div/iframe')
driver.switch_to_frame(iframe)   #切换至登录模块iframe
 
#选择账号密码登录
icon_pc = driver.find_element_by_xpath('/html/body/div/div[2]/div[1]')
icon_pc.click()
time.sleep(2)
name_login = driver.find_element_by_xpath('//*[@id="tab-account"]')
name_login.click()
time.sleep(2)
 
# 输入用户名,密码
username = driver.find_element_by_xpath('//input[@id="account-textbox"]')
password = driver.find_element_by_xpath('//input[@id="password-textbox"]')
username.clear()
username.send_keys(13917444286)
password.clear()
password.send_keys('cpw-118991')
 
    
# 提交登陆
sub_btn = driver.find_element_by_xpath('//button[@id="login-button-account"]')
sub_btn.click()
time.sleep(5)
 
 
while True:
    #检测是否有登录失败警告
    try:
        alert = driver.find_element_by_xpath('/html/body/div/div[1]/div[2]/div[1]/div[3]/span')
    except NoSuchElementException:
        break
    
    if alert:
        #账密登录失败，尝试手机验证码登录
        mobile_login = driver.find_element_by_xpath('//*[@id="tab-mobile"]')
        mobile_login.click()
        username = driver.find_element_by_xpath('//*[@id="mobile-number-textbox"]')
        username.clear()
        username.send_keys(13917444286)
        
        #点击获取验证码
        get_code = driver.find_element_by_xpath('//*[@id="send-number-button"]')
        get_code.click()
        
        #输入验证码
        verify_code = driver.find_element_by_xpath('//*[@id="number-textbox"]')
        verify_code_ = input('verify_code > ')
        verify_code.clear()
        verify_code.send_keys(verify_code_)
        
        # 提交登陆
        sub_btn = driver.find_element_by_xpath('//*[@id="login-button-mobile"]')
        sub_btn.click()
        time.sleep(5)
        break
    else:
        raise Exception("Mobile login failed!")
               
#切换回主页
```
